```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# 生成示例数据
x = [5,10,15,20,30,40,50,70,100]
y1 = [0.4,0.53,0.6,0.55,0.52,0.5,0.45,0.43,0.4]
y2 = [3500,3200,3000,2500,2200,2000,1800,1700,1200]

# 设置图表对象和子图
fig, ax = plt.subplots()

# 绘制第一个数据集
sns.lineplot(x=x, y=y1, ax=ax,color="blue")

# 创建第二个轴对象
ax2 = ax.twinx()

# 绘制第二个数据集
sns.lineplot(x=x, y=y2, ax=ax2,color="red")

# 设置轴标签
# ax.set_ylabel('R2')
# ax2.set_ylabel('Num of Samples')

# 设置图例
ax.legend(labels=['R2'], loc='upper left')
ax2.legend(labels=['Num of Samples'], loc='upper right')
plt.xticks(x)

# 显示图表
plt.show()
```
