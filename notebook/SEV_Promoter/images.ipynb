```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# 生成示例数据
# x = [5,10,15,20,30,40,50,70,100]
# y1 = [0.4416,0.4844,0.65,0.6091,0.5920,0.3789,0.6404,0.3926,0.5855]
# y2 = [4229,3540,3050,2618,2144,1856,1689,1446,1302]

x = [5,10,15,20,30,50,100]
y1 = [0.4416,0.4844,0.65,0.6091,0.5920,0.6404,0.5855]
y2 = [4229,3540,3050,2618,2144,1689,1302]

# 设置图表对象和子图
fig, ax = plt.subplots()

# 绘制第一个数据集
sns.lineplot(x=x, y=y1, ax=ax,color="blue")

# 创建第二个轴对象
ax2 = ax.twinx()

# 绘制第二个数据集
sns.lineplot(x=x, y=y2, ax=ax2,color="red")

# 设置轴标签
# ax.set_ylabel('R2')
# ax2.set_ylabel('Num of Samples')

# 设置图例
ax.legend(labels=['R2'], loc='upper left')
ax2.legend(labels=['Num of Samples'], loc='upper right')
plt.xticks(x)

# 显示图表
plt.show()
```
